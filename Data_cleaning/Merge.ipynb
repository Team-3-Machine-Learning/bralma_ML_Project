Merge to make 1 csv file

```python
import pandas as pd
import glob
import os

# Path to folder containing your CSV files
path = "data/"

# Get all CSV files in the folder
all_files = glob.glob(os.path.join(path, "*.csv"))

# Create an empty list to store dataframes
dfs = []

for file in sorted(all_files):
    df = pd.read_csv(file)
    
    # Optional: add a 'Year' column (if not already in data)
    # You can extract the year from the filename
    year = os.path.basename(file).split('_')[-1].split('.')[0]
    df['Year'] = year
    
    dfs.append(df)

# Combine all into one DataFrame
combined = pd.concat(dfs, ignore_index=True)

# Save as one combined CSV
combined.to_csv("electricity_demand_2001_2025.csv", index=False)

print("✅ Combined file created successfully!")
```
