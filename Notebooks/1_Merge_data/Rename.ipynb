# Renaming columns in dataset

```python
import pandas as pd
import os

path_dataset1 = "../../Dataset1/Merged/electricity_demand_2001_2025.csv"
path_dataset2 = "../../Dataset2/Raw/price_paid_records.csv"
path_dataset2_out = "../../Dataset2/Renamed/price_paid_records_renamed.csv"
```

## Dataset 1 renaming

```python
df1 = pd.read_csv(path_dataset1)
df1

df1.columns = [col.replace(' ', '_') for col in df1.columns]
df1.columns = [col.lower() for col in df1.columns]
df1.to_csv(path_dataset1, index=False)
```

## Dataset 2 renaming

```python
df2 = pd.read_csv(path_dataset2)
df2

df2.columns = [col.replace(' ', '_') for col in df2.columns]
df2.columns = [col.lower() for col in df2.columns]
df2.to_csv(path_dataset2_out, index=False)
```
