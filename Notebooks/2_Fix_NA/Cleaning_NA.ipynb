# Replace with NA

```python
import pandas as pd

path_dataset1 = "../../Dataset1/Merged/electricity_demand_2001_2025.csv"
path_dataset1_out = "../../Dataset1/Cleaned/electricity_demand_2001_2025_NA.csv"

path_dataset2 = "../../Dataset2/Renamed/price_paid_records_renamed.csv"
path_dataset2_out = "../../Dataset2/Cleaned/price_paid_records_NA.csv"
```

## Dataset1 replacing

```python
df1 = pd.read_csv(path_dataset1)
df1 = df1.fillna('NA')
df1.to_csv(path_dataset1_out, index=False)
```

## Dataset2 replacing

```python
df2 = pd.read_csv(path_dataset2)
df2 = df2.fillna('NA')
df2.to_csv(path_dataset2_out, index=False)
```
