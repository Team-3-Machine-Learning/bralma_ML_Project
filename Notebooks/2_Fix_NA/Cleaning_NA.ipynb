```python
# Replace NaN or missing values with 'NA' in the dataset
import pandas as pd

# Load the dataset
file_path = '../../Dataset1/Merged/electricity_demand_2001_2025.csv'
data = pd.read_csv(file_path)

# Replace NaN or missing values with 'NA', but keep 0 values intact
data = data.fillna('NA')

# Save the cleaned dataset
output_path = '../../Dataset1/Cleaned/electricity_demand_2001_2025_NA.csv'
data.to_csv(output_path, index=False)
print(f"Cleaned data saved to {output_path}")
```
