# Dataset cleaning

```python
!pip3 install pandas numpy
import pandas as pd
import numpy as np

path_dataset1 = "../../Dataset1/Merged/electricity_demand_2001_2025.csv"
path_dataset1_out = "../../Dataset1/Cleaned/electricity_demand_2001_2025_mean.csv"

path_dataset2 = "../../Dataset2/Renamed/price_paid_records_renamed.csv"
path_dataset2_out = "../../Dataset2/Cleaned/price_paid_records_mean.csv"
```

## Dataset 1 cleaning

```python
df1 = pd.read_csv(path_dataset1)
na_columns = df1.columns[df1.isna().any()].tolist()

means1 = df1.mean(numeric_only=True)

df1_filled = df1.fillna(means1)
df1_filled.to_csv(path_dataset1_out, index=False)
```

## Dataset2 cleaning

```python
df2 = pd.read_csv(path_dataset2)
na_columns = df2.columns[df2.isna().any()].tolist()

means2 = df2.mean(numeric_only=True)

df2_filled = df2.fillna(means2)
df2_filled.to_csv(path_dataset2_out, index=False)
```
