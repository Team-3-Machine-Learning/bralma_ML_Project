```python
# Load data
df = pd.read_csv("../../Dataset1/Merged/electricity_demand_2001_2025.csv")
df
```

```python
# Checked welke kolommen NA-values hebben
na_columns = df.columns[df.isna().any()].tolist()
print("Kolommen met NA waarden:", na_columns)
```

```python
# Bereken de mean voor elke kolom
means = df.mean(numeric_only=True)
print(means)
```

```python
df_filled = df.fillna(means)
df_filled.to_csv("../../Dataset1/Cleaned/electricity_demand_2001_2025_mean.csv", index=False)
print("✅ Bestand opgeslagen als electricity_demand_2001_2025_mean.csv")
```
