```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "../../Dataset1/Merged/electricity_demand_2001_2025.csv"
df = pd.read_csv(path)
df
```

```python
(df == 0).sum()
# Controleer hoeveel nullen er zijn per kolom
zero_counts = (df == 0).sum().sort_values(ascending=False)
print(zero_counts.head(10))

# Controleer algemene beschrijving
df.describe()
```

```python
import matplotlib.pyplot as plt

for col in ['embedded_wind_generation', 'embedded_solar_generation']:
    plt.figure()
    df[col].plot(kind='hist', bins=50, title=col)
    plt.show()
```

```python
# === Behandeling van nulwaarden ===

# 1. embedded_wind_generation → 0 = geen meting vóór 2005
df.loc[(df['year'] < 2005) & (df['embedded_wind_generation'] == 0), 'embedded_wind_generation'] = np.nan

# 2. embedded_solar_generation → 0 = geen meting vóór 2011
df.loc[(df['year'] < 2011) & (df['embedded_solar_generation'] == 0), 'embedded_solar_generation'] = np.nan

# 3. non_bm_stor → bevat bijna alleen nullen, verwijderen
if (df['non_bm_stor'] == 0).mean() > 0.95:
    df.drop(columns=['non_bm_stor'], inplace=True)

# 4. NaN’s invullen door interpolatie (aangezien het tijdsreeksdata is)
df['embedded_wind_generation'] = df['embedded_wind_generation'].interpolate()
df['embedded_solar_generation'] = df['embedded_solar_generation'].interpolate()
```
